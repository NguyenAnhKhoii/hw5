Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 

```python
def check_armstrong(n):
    k = len(str(n))
    temp = n
    s = 0
    
    while temp > 0:
        digit = temp % 10
        s += digit ** k
        temp //= 10
    
    return s == n
    print(is_armstrong(153))
    print(is_armstrong(123))
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 


```python
def check_square(n):
    i = 1
    while i * i < n:
        i += 1
    return i * i == n
print(check_square(9))  
print(check_square(10)) 
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
def check_palindrome(n):
    temp = n
    reverse = 0
    
    while temp > 0:
        digit = temp % 10
        reverse = reverse * 10 + digit
        temp //= 10
    
    return reverse == n
print(check_palindrome(121))
print(check_palindrome(123))
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
def check_magic(n):
    while n > 9:   # còn nhiều hơn 1 chữ số
        s = 0
        while n > 0:
            s += n % 10
            n //= 10
        n = s
    return n == 1
print(check_magic(19)) 
print(check_magic(123)) 
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.


```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except:
        try:
            float(s)
            return 'float'
        except:
            return 'neither'
print(detect_number_type("123"))
print(detect_number_type("12.3"))
print(detect_number_type("abc"))
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        return 1 if b != 0 else 0
    delta = b*b - 4*a*c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1)*(x3 - x1) == (y3 - y1)*(x2 - x1)
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: 

```python
def area(a, b, c):
    return abs((a[0]*(b[1]-c[1]) + b[0]*(c[1]-a[1]) + c[0]*(a[1]-b[1])) / 2)

def point_in_quadrilateral(quad, p):
    A, B, C, D = quad
    S = area(A,B,C) + area(A,C,D)
    S1 = area(p,A,B) + area(p,B,C) + area(p,C,D) + area(p,D,A)
    return abs(S - S1) < 1e-6
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_nonzero_digits(lst):
    count = 0
    for num in lst:
        for digit in str(abs(num)):
            if digit != '0':
                count += 1
    return count
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math

def angle_ABC(x1,y1,x2,y2,x3,y3):
    BA = (x1-x2, y1-y2)
    BC = (x3-x2, y3-y2)
    
    dot = BA[0]*BC[0] + BA[1]*BC[1]
    normBA = math.sqrt(BA[0]**2 + BA[1]**2)
    normBC = math.sqrt(BC[0]**2 + BC[1]**2)
    
    cos_angle = dot / (normBA * normBC)
    angle = math.degrees(math.acos(cos_angle))
    return angle
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
import math

def C(n, k):
    return math.factorial(n) // (math.factorial(k) * math.factorial(n-k))
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm v

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    base = 10000
    per_km = 12000 if car_type == "4" else 15000
    wait_fee = 2000 * (wait_minutes // 5)
    total = base + km * per_km + wait_fee
    if start_hour >= 22 or start_hour < 5:
        total *= 1.1
    if km > 20:
        total *= 0.9    
    return int(total)
```
